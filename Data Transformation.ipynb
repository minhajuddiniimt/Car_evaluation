```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('car.csv',names=['buying','maint','doors','person','lug_boot','safty','class'])
```

```python
data.head()
```

### Attribute Buying
we crete dummies of 'buying' attribute and add in data

```python
buying = pd.get_dummies(data['buying'])
```

```python
buying = buying.drop('vhigh', axis=1)
```

```python
print(buying.shape)
buying.head()
```

```python
data[['buy_high','buy_low','buy_med']]= buying[['high','low','med']]
```

```python
data.head()
```

### Similar maint, doors, person, lug_boot, safty

```python
maint = pd.get_dummies(data['maint']).drop('vhigh',axis=1)
doors = pd.get_dummies(data['doors']).drop('5more', axis=1)
person = pd.get_dummies(data['person']).drop('more',axis=1)
lug_boot = pd.get_dummies(data['lug_boot']).drop('big',axis=1)
safty = pd.get_dummies(data['safty']).drop('high', axis=1)
```

```python
data[['maint_low','maint_med','maint_high']]= maint[['low','med','high']]
```

```python
data.head()
```

```python
data[['doors_2','doors_3','doors_4']]=doors[['2','3','4']]
```

```python
data[['person_2','person_4']]=person[['2','4']]
```

```python
data[['lug_boot_med','lug_boot_small']]= lug_boot[['med','small']]
```

```python
data[['safty_med','safty_low']]=safty[['med','low']]
```

```python
data.head()
```

```python
data.head()
```

```python
car_class = pd.get_dummies(data['class'])
```

```python
data[['acc','good','unacc','vgood']]=car_class[['acc','good','unacc','vgood']]
```

### Drop unwanted columns

```python
data = data.drop(['buying','maint','doors','person','lug_boot','safty'], axis=1)
```

```python
data = data.drop('class',axis=1)
```

### Export the data

```python
data.to_csv('Transform_data.csv', index=False)
```
